```
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```
from google.colab import drive
import os
import cv2
from skimage.transform import rescale
drive.mount('/content/drive')

def load_images(path, label):
  img_arr = []
  for filename in os.listdir(path):
    # We only need to keep 1 color channel, since the images have been grayscaled
    img = cv2.imread(path + filename)[:,:,0]

    # Downscale since we keep running out of RAM
    img = rescale(img, 0.25, anti_aliasing=True)

    # Data is formatted as [image,label]
    img_arr.append([img,label]) 
  return img_arr

train_path = '/content/drive/MyDrive/ML Final Project/ML Final Project Data/Processed/Train/'
val_path = '/content/drive/MyDrive/ML Final Project/ML Final Project Data/Processed/Val/'

proc_waffles_path = train_path + 'waffles/'
proc_sushi_path = train_path + 'sushi/'
proc_wild_path = train_path + 'wild/'
proc_cat_path = train_path + 'cat/'
proc_takoyaki_path = train_path + 'takoyaki/'
proc_tacos_path = train_path + 'tacos/'
proc_faces_path = train_path + 'faces/'
proc_tiramisu_path = train_path + 'tiramisu/'
proc_tuna_tartare_path = train_path + 'tuna_tartare/'
proc_dog_path = train_path + 'dog/'

npy_path = '/content/drive/MyDrive/ML Final Project/npy files'

# Pass in a label as well:
# Waffles - 0
# Sushi - 1
# Wild - 2
# Cat - 3
# Takoyaki - 4
# Tacos - 5
# Faces - 6
# Tiramisu - 7
# Tuna Tartare - 8
# Dog - 9
waffles_dat = load_images(proc_waffles_path, 0)
np.save(os.path.join(npy_path,'waffles_dat'), waffles_dat, allow_pickle=True)
sushi_dat = load_images(proc_sushi_path, 1)
np.save(os.path.join(npy_path,'sushi_dat'), sushi_dat, allow_pickle=True)
wild_dat = load_images(proc_wild_path, 2)
np.save(os.path.join(npy_path,'wild_dat'), wild_dat, allow_pickle=True)
cat_dat = load_images(proc_cat_path, 3)
np.save(os.path.join(npy_path,'cat_dat'), cat_dat, allow_pickle=True)
takoyaki_dat = load_images(proc_takoyaki_path, 4)
np.save(os.path.join(npy_path, 'takoyaki_dat'), takoyaki_dat, allow_pickle=True)
tacos_dat = load_images(proc_tacos_path, 5)
np.save(os.path.join(npy_path, 'tacos_dat'), tacos_dat, allow_pickle=True)
faces_dat = load_images(proc_faces_path, 6)
np.save(os.path.join(npy_path, 'faces_dat'), faces_dat, allow_pickle=True)
tiramisu_dat = load_images(proc_tiramisu_path, 7)
np.save(os.path.join(npy_path, 'tiramisu_dat'), tiramisu_dat, allow_pickle=True)
tuna_tartare_dat = load_images(proc_tuna_tartare_path, 8)
np.save(os.path.join(npy_path, 'tuna_tartare_dat'), tuna_tartare_dat, allow_pickle=True)
dog_dat = load_images(proc_dog_path, 9)
np.save(os.path.join(npy_path, 'dog_dat'), dog_dat, allow_pickle=True)
```

```
# Convert to np array
waffles_dat = np.array(waffles_dat)
sushi_dat = np.array(sushi_dat)
wild_dat = np.array(wild_dat)
cat_dat = np.array(cat_dat)
takoyaki_dat = np.array(takoyaki_dat)
tacos_dat = np.array(tacos_dat)
faces_dat = np.array(faces_dat)
tiramisu_dat = np.array(tiramisu_dat)
tuna_tartare_dat = np.array(tuna_tartare_dat)
dog_dat = np.array(dog_dat)

print("Waffles: ", np.shape(waffles_dat))
print("Sushi: ", np.shape(sushi_dat))
print("Wild:   ", np.shape(wild_dat))
print("Cat:  ", np.shape(cat_dat))
print("Takoyaki: ", np.shape(takoyaki_dat))
print("Tacos:  ", np.shape(tacos_dat))
print("Faces:  ", np.shape(faces_dat))
print("Tiramisu:  ", np.shape(tiramisu_dat))
print("Tuna Tartare: ", np.shape(tuna_tartare_dat))
print("Dog:   ", np.shape(dog_dat))

# Show an example
img = waffles_dat[0][0]
print("Image dimensions: ", np.shape(img))
plt.figure()
plt.imshow(img, cmap="gray")
```

```
from google.colab import drive
import os
import cv2
drive.mount('/content/drive')
npy_path = '/content/drive/MyDrive/ML Final Project/npy files'



# Concatenate together
waffles_dat = np.load(os.path.join(npy_path,'waffles_dat.npy'), allow_pickle=True)
sushi_dat = np.load(os.path.join(npy_path,'sushi_dat.npy'), allow_pickle=True)
wild_dat = np.load(os.path.join(npy_path,'wild_dat.npy'), allow_pickle=True)
cat_dat = np.load(os.path.join(npy_path,'cat_dat.npy'), allow_pickle=True)
takoyaki_dat = np.load(os.path.join(npy_path,'takoyaki_dat.npy'), allow_pickle=True)
tacos_dat = np.load(os.path.join(npy_path,'tacos_dat.npy'), allow_pickle=True)
faces_dat = np.load(os.path.join(npy_path,'faces_dat.npy'), allow_pickle=True)
tiramisu_dat = np.load(os.path.join(npy_path,'tiramisu_dat.npy'), allow_pickle=True)
tuna_tartare_dat = np.load(os.path.join(npy_path,'tuna_tartare_dat.npy'), allow_pickle=True)
dog_dat = np.load(os.path.join(npy_path,'dog_dat.npy'), allow_pickle=True)

all_data = np.concatenate((waffles_dat, sushi_dat, wild_dat, cat_dat, takoyaki_dat, tacos_dat, faces_dat, tiramisu_dat, tuna_tartare_dat, dog_dat))

# Shuffle the data
np.random.shuffle(all_data)
# Split images and labels apart
inputs = all_data[:,0]
y_labels = all_data[:,1]
np.save(os.path.join(npy_path,'inputs'), inputs, allow_pickle=True)
np.save(os.path.join(npy_path ,'labels'), y_labels, allow_pickle=True)

print("Input size: ", np.shape(inputs))
print("Label size: ", np.shape(y_labels))

# Show an example
img = inputs[0]
print("Image dimensions: ", np.shape(img))
plt.figure()
plt.imshow(img, cmap="gray")
plt.title(f"Label {y_labels[0]}")
```
